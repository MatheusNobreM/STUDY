```python
import re
```

```python
def explorar_expressa_regular(texto):
  padrao = r'\bPython\b'
  if re.search(padrao, texto):
    print("O padrão foi encontrado.")
  else:
    print("O padrão não for encontrado.")
  
  padrao_email = r'[a-zA-Z0-9.%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}'
  email = re.findall(padrao_email, texto)
  print(f'E-mail encontrado: {email}')
  
  texto_sub = re.sub(r'Python', 'Java', texto)
  print(texto_sub)
  
  texto_dividido = re.split(r'\s+', texto)
  print(texto_dividido) 
```

```python
texto = "Olá sou programdo Python, exemplo@dominio.com"
explorar_expressa_regular(texto)
```
