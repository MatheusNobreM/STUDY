```python
numeros = range(6)
```

```python
def celsius_para_fahrenheit(c):
  return (c * 9/5) + 32
```

```python
temperaturas_celsius = [0,10,20,30,40]
temperaturas_fahrenheit = map (celsius_para_fahrenheit, temperaturas_celsius)
```

```python
list(temperaturas_fahrenheit)
```
