```python
def e_par(x):
  return x % 2 == 0
```

```python
numeros = range(10)
```

```python
numeros_pares = filter(e_par, numeros)
list(numeros_pares)
```

```python
num_map = map(e_par, numeros)
list(num_map)
```

```python
produtos = [
  {"nome": "A", "em_estoque": True},
  {"nome": "B", "em_estoque": False},
  {"nome": "C", "em_estoque": False},
  {"nome": "D", "em_estoque": True},

]
```

```python
def produto_em_estoque(produto):
  return produto["em_estoque"]
```

```python
produto_disponiveis = filter(produto_em_estoque, produtos)
list(produto_disponiveis)
```
