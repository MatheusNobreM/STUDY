```python
def linhas_com_numeros_de_palavras(arquivo, numero):
  with open(arquivo, 'r', encoding='utf-8') as file:
    print(file)
    for linha in file:
      if len(linha.split()) == numero:
        yield linha.strip()
```

```python
arquivo = 'txt.txt'
numero_palavras = 2
for linha in linhas_com_numeros_de_palavras(arquivo, numero_palavras):
  print(linha)
```
