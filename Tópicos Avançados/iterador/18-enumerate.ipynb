```python
frutas = ["Maça", "banana", "laranja"]
```

```python
for fruta in frutas:
  print(fruta)
```

```python
for fruta in enumerate(frutas):
  print(fruta)
```

```python
for indice, fruta in enumerate(frutas):
  print(f"{indice}-{fruta}")
```

```python
num = [10, 20, 30, 40, 50]
```

```python
indice_valores = {indice: valor for indice, valor in enumerate(num)}
indice_valores
```

```python
indice_par = {indice: valor for indice, valor in indice_valores.items() if indice % 2 == 0}
indice_par
```
