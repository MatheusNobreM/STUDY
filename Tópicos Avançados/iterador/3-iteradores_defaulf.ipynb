```python
import itertools
```

```python
for i in range(5):
  print(i)
```

```python
frutas = ['maça', 'banana', 'limao']
for fruta in frutas:
  print(fruta)
```

```python
pessoas = {'nome': 'alice', 'idade': 30}

for chave in pessoas:
  print(chave, pessoas[chave])
```

```python
for indice, fruta in enumerate(frutas):
  print(indice, fruta)
```

```python
nomes = ['alice', 'bob', 'carlos']
idades = [60, 12, 4]

for nome, idade in zip(nomes, idades):
  print(nome, idade)
```

```python
for i in itertools.count(start=10, step=2):
  if i>20:
    break
  print(i)
```
