```python
from typing import List
```

```python
class Estudante:
  def __init__(self, nome: str, notas: List[float]) -> None:
    self.nome = nome
    self.notas = notas
    
  def media(self) -> float:
    return sum(self.notas) / len(self.notas)
```

```python
def alunos_acima_da_media(alunos : List[Estudante], media_minima: float) -> List[str]:
  return [aluno.nome for aluno in alunos if aluno.media() >= media_minima]
```

```python
alunos = [
  Estudante("Alice", [7.6, 8, 3.4]),
  Estudante("Matheus", [7.6, 8, 9]),
  Estudante("Jose", [2.3, 8, 3.4]),
  Estudante("Alle", [8.2, 4.9, 10])
]
```

```python
resultados = alunos_acima_da_media(alunos, 7)
resultados
```
