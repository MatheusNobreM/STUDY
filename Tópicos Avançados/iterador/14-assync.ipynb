```python
import asyncio
import aiohttp
```

```python
async def tarefa(nome, tempo):
  print(f"Tarefa {nome} iniciada.")
  await asyncio.sleep(tempo)
  print(f"Tarefa {nome} concluida.")
```

```python
async def main():
  await asyncio.gather(
    tarefa("A", 2),
    tarefa("B", 1)
  )
```

```python
try:
  asyncio.get_running_loop()
except RuntimeError:
  asyncio.run(main())
```

```python
asyncio.ensure_future(main())
```

```python
async def buscar_usuario(session, usuario_id):
  async with session.get(
    f'https://jsonplaceholder.typicode.com/users/{usuario_id}'
  ) as resposta:
    return await resposta.json()
```

```python
async def main():
  async with aiohttp.ClientSession() as session:
    usuario_id = range(6)
    tarefas = [buscar_usuario(session, uid) for uid in usuario_id]
    
    #Aguarda a conclus√£o de toas as tarefas
    usuarios = await asyncio.gather(*tarefas)
    for usuario in usuarios:
      print(usuario)
```

```python
try:
  asyncio.get_running_loop()
except RuntimeError:
  asyncio.run(main())
```

```python
asyncio.ensure_future(main())
```
